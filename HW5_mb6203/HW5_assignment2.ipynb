```python
SQL_SOURCE = 'https://fb55.carto.com/api/v2/sql?q=SELECT'

import urllib2
import urllib
import StringIO
import ast
import pandas as pd

try:
    from StringIO import BytesIO as io
except ImportError:
    from io import BytesIO as io

def queryCartoDB(query, format='CSV', source=SQL_SOURCE):
    '''queries carto datasets from a given carto account
    Arguments: 
    query - string: a valid sql query string
    format - outlut format  OPTIONAL (default CSV)
    source - a valid sql api endpoint OPTIONAL (default carto fb55 account)
    Returns:
    the return of the sql query AS A STRING
    NOTES:
    designed for the carto API, tested only with CSV return format'''
    
    data = urllib.urlencode({'format': format, 'q': query})
    try:
        response = urllib2.urlopen(source, data)
    except urllib2.HTTPError, e:
        raise ValueError('\n'.join(ast.literal_eval(e.readline())['error']))
    except Exception:
        raise
    return response.read()

def get_data(query):
    try:
        return pd.read_csv(io(queryCartoDB(query)), sep = ',')
    except ValueError as v:
        print (str(v))
```

```python
query = '''SELECT start_station_name,
COUNT(start_station_latitude) as Num_trips_started_per_station 
FROM citibike
GROUP BY start_station_name;'''

get_data(query)
```

```python
#Showing the number of trips started per station, but only for stations within 500m of Time Square! 
#The coordinates of Time Square is (40.7577,-73.9857)
query = '''
SELECT Count(start_station_name) as num_trips_per_station, start_station_name FROM citibike 
WHERE 
ROUND(6378.138 * 2 * ASIN(SQRT(POW(SIN((40.7577 * PI() / 180 - start_station_latitude * PI() / 180) / 2),2) 
+ COS(40.7577 * PI() / 180) * COS(start_station_latitude * PI() / 180) * 
POW(SIN((-73.9857 * PI() / 180 - start_station_longitude * PI() / 180) / 2),2))) * 1000)
<500
GROUP BY start_station_name
'''



get_data(query)
```

```python
query = '''
SELECT avg(tripduration) AS av_tripduration, Count(start_station_name) as num_trips_per_station, start_station_name, SUM(tripduration) AS trip_duration
FROM citibike 

WHERE 
ROUND(6378.138 * 2 * ASIN(SQRT(POW(SIN((40.7577 * PI() / 180 - start_station_latitude * PI() / 180) / 2),2) 
+ COS(40.7577 * PI() / 180) * COS(start_station_latitude * PI() / 180) * 
POW(SIN((-73.9857 * PI() / 180 - start_station_longitude * PI() / 180) / 2),2))) * 1000)
<500 AND EXTRACT(DOW FROM starttime::DATE) IN (0, 6)
GROUP BY start_station_name
ORDER BY av_tripduration DESC
'''



get_data(query)

```
